```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE
from tensorflow.keras.utils import Sequence
from imblearn.over_sampling import RandomOverSampler
from sklearn.utils import shuffle

import numpy as np
```

```python
# Load the pre-saved features and labels from the .npz file
data = np.load('../data/features_and_labels.npz')
X = data['features']  # Extract the features
y = data['labels']    # Extract the labels
```

```python
# Ensure labels are integers
if y.dtype.kind not in {'i', 'u'}:  # Check if labels are not integers
    unique_labels = np.unique(y)    # Find unique label values
    label_map = {label: idx for idx, label in enumerate(unique_labels)}
    y = np.array([label_map[label] for label in y])
```

```python
# Normalize the features
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Encode moods (labels) as integers
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
# Address class imbalance using class weights
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights = dict(enumerate(class_weights))
```

```python
# Load VGG16 without the top classification layer
base_model = tf.keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Freeze the base model layers

num_classes = 7

# Add custom layers after VGG16
model = Sequential()
model.add(base_model)
model.add(Dropout(0.5))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(tf.keras.layers.Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(tf.keras.layers.Dense(num_classes, activation='softmax'))  # Output layer with 'num_classes' classes

# Compile the model with Adam optimizer and sparse categorical crossentropy loss
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Define the RandomOverSampler
ros = RandomOverSampler()

# Create the training data generator with oversampling
train_seq = OversampleSequence(X_train, y_train, batch_size=32, oversampler=ros)

# Create a learning rate scheduler function
def scheduler(epoch, lr):
    if epoch % 10 == 0 and epoch != 0:
        lr = lr * 0.5  # Reduce learning rate by half every 10 epochs
    return lr

lr_scheduler = tf.keras.callbacks.LearningRateScheduler(scheduler)

# Define early stopping and model checkpoint callbacks
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = tf.keras.callbacks.ModelCheckpoint('best_model.keras', save_best_only=True, monitor='val_accuracy')

# Fit the model with the training data generator and validation data
history = model.fit(
    X_train, y_train,  # Use the custom sequence for training
    epochs=100,
    validation_data=(X_test, y_test),  # Validation dataset
    steps_per_epoch=len(train_seq),  # Number of steps per epoch
    callbacks=[early_stopping, checkpoint, lr_scheduler]  # Add callbacks
)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, LearningRateScheduler
from imblearn.over_sampling import RandomOverSampler
from tensorflow.keras.utils import Sequence
import numpy as np

# Define the OversampleSequence class for custom data generator
class OversampleSequence(Sequence):
    def __init__(self, X, y, batch_size, oversampler, target_size=(224, 224)):
        self.X = X
        self.y = y
        self.batch_size = batch_size
        self.oversampler = oversampler
        self.target_size = target_size

    def __len__(self):
        return int(np.ceil(len(self.X) / self.batch_size))

    def __getitem__(self, idx):
        start = idx * self.batch_size
        end = min((idx + 1) * self.batch_size, len(self.X))  # Avoid out-of-bounds
        X_batch, y_batch = self.X[start:end], self.y[start:end]
        
        # Perform oversampling
        X_oversampled, y_oversampled = self.oversampler.fit_resample(X_batch, y_batch)
        
        # Ensure X_oversampled is in the shape (batch_size, height, width, channels)
        if len(X_oversampled.shape) == 3:  # If shape is (height, width, channels)
            X_oversampled = np.expand_dims(X_oversampled, axis=0)  # Add batch dimension
        
        # Ensure that X_oversampled has 4 dimensions: (batch_size, height, width, channels)
        if len(X_oversampled.shape) == 3:
            X_oversampled = np.expand_dims(X_oversampled, axis=0)  # Add batch dimension
            
        # Resize images to target size (224x224)
        X_oversampled_resized = np.array([tf.image.resize(image, self.target_size).numpy() for image in X_oversampled])
        
        # Convert to TensorFlow tensors
        X_oversampled_resized = tf.convert_to_tensor(X_oversampled_resized, dtype=tf.float32)
        y_oversampled = tf.convert_to_tensor(y_oversampled, dtype=tf.int64)
        
        return X_oversampled_resized, y_oversampled


# Load VGG16 without the top classification layer
base_model = tf.keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Freeze the base model layers

num_classes = 7

# Add custom layers after VGG16
model = models.Sequential()
model.add(base_model)
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(num_classes, activation='softmax'))  # Output layer with 'num_classes' classes

# Compile the model with Adam optimizer and sparse categorical crossentropy loss
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Define the RandomOverSampler
ros = RandomOverSampler()

# Create the training data generator with oversampling
train_seq = OversampleSequence(X_train, y_train, batch_size=32, oversampler=ros)

# Create a learning rate scheduler function
def scheduler(epoch, lr):
    if epoch % 10 == 0 and epoch != 0:
        lr = lr * 0.5  # Reduce learning rate by half every 10 epochs
    return lr

lr_scheduler = LearningRateScheduler(scheduler)

# Define early stopping and model checkpoint callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', save_best_only=True, monitor='val_accuracy')

# Fit the model with the training data generator and validation data
history = model.fit(
    train_seq,  # Use the custom sequence for training
    epochs=100,
    validation_data=(X_test, y_test),  # Validation dataset
    steps_per_epoch=len(train_seq),  # Number of steps per epoch
    callbacks=[early_stopping, checkpoint, lr_scheduler]  # Add callbacks
)
```
